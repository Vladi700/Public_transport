```python
nodes = pd.read_csv("cities/tokyo_nodes.csv")
edges = pd.read_csv("cities/tokyo_edges.csv")

# Keep first occurrence only
nodes = nodes.drop_duplicates(subset=["nodeID"], keep="first")
node_ids = nodes["nodeID"].astype(int).tolist()
coords = nodes.set_index("nodeID")[["latitude", "longitude"]].to_dict(orient="index")

n = len(node_ids)
adj = [dict() for _ in range(n+1)]

for _, row in edges.iterrows():
    u_id = int(row["nodeID_from"])
    v_id = int(row["nodeID_to"])


    cu, cv = coords[u_id], coords[v_id]
    w = _haversine_km(cu["latitude"], cu["longitude"], cv["latitude"], cv["longitude"])

    adj[u_id][v_id] = w
    adj[v_id][u_id] = w


```

```python
bmssp_graph = Bmssp(adj)
res_0 = bmssp_graph.solve(origin_id=283)
```

```python
def is_connected(adj):
    visited = set()
    stack = [0]
    while stack:
        u = stack.pop()
        if u in visited:
            continue
        visited.add(u)
        stack.extend(adj[u].keys())
    return len(visited), len(adj)

print(is_connected(adj))
```
